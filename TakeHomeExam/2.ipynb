# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Given Data

```python
t = np.array([-2, -1, 0, 1, 2])
y = np.array([2, -10, 0, 2, 1])
```

# Given Function

```python
def p(t, x0, x1, x2):
    return x0 + x1 * t + x2 * t ** 2
```

# Cost Function 

```python
def cost_function(t, y, x0, x1, x2):
    N = len(t)
    squared_errors = [(p(ti, x0, x1, x2) - yi) ** 2 for ti, yi in zip(t, y)]
    return sum(squared_errors)
```

# Gradient Descent

```python
def gradient_descent(t, y, learning_rate, iterations):
    x0 = 0
    x1 = 0
    x2 = 0
    
    for _ in range(iterations):
        grad_x0 = sum([p(ti, x0, x1, x2) - yi for ti, yi in zip(t, y)])
        grad_x1 = sum([(p(ti, x0, x1, x2) - yi) * ti for ti, yi in zip(t, y)])
        grad_x2 = sum([(p(ti, x0, x1, x2) - yi) * (ti ** 2) for ti, yi in zip(t, y)])
        
        x0 -= learning_rate * grad_x0
        x1 -= learning_rate * grad_x1
        x2 -= learning_rate * grad_x2
        
    return x0, x1, x2
```

# Main

```python
learning_rate = 0.01
iterations = 1000
x0_opt, x1_opt, x2_opt = gradient_descent(t, y, learning_rate, iterations)

print("Optimal parameters:")
print("x0:", x0_opt)
print("x1:", x1_opt)
print("x2:", x2_opt)
```

# Plotting

```python
plt.figure(figsize=(12, 8))

t_fit = np.linspace(min(t), max(t), 100)
y_fit = [p(ti, x0_opt, x1_opt, x2_opt) for ti in t_fit]

# Plot the data and the fitted polynomial
plt.scatter(t, y, label='Data points')
plt.plot(t_fit, y_fit, color='red', label='Fitted polynomial')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Fitted polynomial vs Data points')
plt.legend()
plt.grid(True)
plt.savefig("./Images/Problem-2-cost-vs-iter.pdf")
plt.show()
```
