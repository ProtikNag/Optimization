# Given Function

```python
def f(X, x0, x1, x2, x3):
    return x0 + x1*(np.sin(X) + np.cos(X)) + x2*(np.sin(2*X) + np.cos(2*X)) + x3*(np.sin(3*X) + np.cos(3*X))
```

# Cost Function 

```python
def cost_function(X, Y, x0, x1, x2, x3):
    m = len(X)
    predicted = f(X, x0, x1, x2, x3)
    return (1/(2*m)) * np.sum((predicted - Y)**2)
```
