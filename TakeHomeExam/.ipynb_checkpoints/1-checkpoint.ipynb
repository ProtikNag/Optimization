# Import 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
```

# Objective Function

```python
def f(x):
    x1, x2, x3 = x
    
    return 5*x1**2 + x2**2 + 2*x3**2 + 4*x1*x2 - 14*x1 - 6*x2 + 20
```

# Gradient of Objective Function

```python
def grad_f(x):
    x1, x2, x3 = x
    df_dx1 = 10*x1 + 4*x2 - 14
    df_dx2 = 2*x2 + 4*x1 - 6
    df_dx3 = 4*x3
    
    return np.array([df_dx1, df_dx2, df_dx3])
```

# Step Size Selection

### Minimization

```python
def minimize_step_size(gradient):
    # calculation is presented in the pdf 
    nominator = gradient[0]**2 + gradient[1]**2 + gradient[2]**2
    denominator = 5*gradient[0]**2 + gradient[1]**2 + 2*gradient[2]**2 + 4*gradient[0]*gradient[1]
    alpha = nominator / (2 * denominator)

    return alpha
```

### Bisection

```python
def bisection_step_size(x, gradient):
    direction = -gradient
    alpha_lower = 0
    alpha_upper = 1
    epsilon = 1e-6
    while alpha_upper - alpha_lower > epsilon:
        alpha = (alpha_lower + alpha_upper) / 2
        h_prime_alpha = np.dot(grad_f(x + alpha * direction), direction)
        if h_prime_alpha < 0:
            alpha_lower = alpha
        else:
            alpha_upper = alpha
    return (alpha_lower + alpha_upper) / 2
```

# Steepest Descent 

```python
def steepest_descent(x, step_size_rule, max_iter, tol):
    history = []
    history.append((x, f(x)))

    for i in range(max_iter):
        grad = grad_f(x)
        
        if step_size_rule == 'minimization':
            lr = minimize_step_size(grad)
        elif step_size_rule == 'bisection':
            lr = bisection_step_size(x, grad)
        elif step_size_rule == 'diminishing':
            lr = 0.01 / np.sqrt(i + 1)
        elif step_size_rule == 'constant':
            lr = 0.01

        x_new = x - lr * grad
        history.append((x_new, f(x_new)))
        
        if np.linalg.norm(grad) < tol:
            break
            
        x = x_new
        
    return x, history
```

# Main

```python
x0 = np.array([1, 5, 0])
max_iter = 50
tol = 1e-6                                 # tolerance

step_size_rules = ['minimization', 'bisection', 'diminishing', 'constant']

plt.figure(figsize=(12, 8))

for rule in step_size_rules:
    x, history = steepest_descent(x0, rule, max_iter, tol)
    print(f"Step Size Rule: {rule} ---- Optimized Value: {x}")
    iterations = np.arange(len(history))
    costs = [cost for _, cost in history]
    plt.plot(iterations, costs, label=rule)

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost vs Iteration (Steepest Descent with Different Step Size Rules)')
plt.legend()
plt.grid(True)
plt.savefig("./Images/Problem-1-cost-vs-iteration.pdf")
plt.show()
```
