# Import 

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Plotting

```python
# Define the function
def objective_function(x1, x2):
    return (x1 - 3)**2 + (x2 - 3)**2

# Constraint 1: 4(x1)^2 + 9(x2)^4 <= 36
x1 = np.linspace(-3, 3, 400)
x2_upper = np.sqrt((36 - 4*x1**2) / 9)
x2_lower = -x2_upper

def ellipse(x1, x2):
  return x1**2 + 3 * x2**2 - 3

# Create mesh grid
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-2, 2, 400)
X1, X2 = np.meshgrid(x1, x2)

# Evaluate equation on the grid
Z = ellipse(X1, X2)

plt.figure(figsize=(12, 8))
# Create the plot
plt.contour(X1, X2, Z, levels=[0])
# Plotting

plt.plot(x1, x2_upper, 'b', label='4(x1)^2 + 9(x2)^4 <= 36')
plt.plot(x1, x2_lower, 'b')
# plt.plot(x1, x2_ellipse, 'r', label='(x1)^2 + 3(x2)^2 = 3')
# plt.plot(x1, -x2_ellipse, 'r')

# plt.show()

# Define range for x1 and x2
x1 = np.linspace(-3, 6, 400)
x2 = np.linspace(-3, 6, 400)

# Create a grid of x1 and x2 values
X1, X2 = np.meshgrid(x1, x2)

# Compute the function values for the grid points
Z = objective_function(X1, X2)

plt.contour(X1, X2, Z, levels=np.linspace(0, 20, 21), cmap='viridis')

x1 = np.linspace(-10, 10, 1000)
x2 = np.linspace(-10, 10, 1000)

# Create a grid of x1 and x2 values
X1, X2 = np.meshgrid(x1, x2)

a = [0.9]
b = [0.85]

plt.fill_betweenx(x2, -1, 3, where=(x2 >= -3), color='lightblue', label='x1 >= -1')
# plt.scatter(a, b, 'x')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Feasible Region')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.savefig("./Images/Problem-4-feasible-region.pdf")

plt.show()
```
