```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize
```

```python
def objective(x):
    return (x[0] - 3)**2 + (x[1] - 3)**2

def constraint_inequality(x):
    return 4*x[0]**2 + 9*x[1]**2 - 36

def constraint_equality(x):
    return x[0]**2 + 3*x[1] - 3

def constraint_feasible(x):
    return x[0] + 1  # x1 >= -1

x1 = np.linspace(-4, 4, 100)
x2 = np.linspace(-4, 4, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = (X1 - 3)**2 + (X2 - 3)**2

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(15, -30)
ax.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.9, rstride=100, cstride=100, edgecolor='k')
ax.plot_surface(X1, X2, np.zeros_like(Z), color='gray', alpha=0.5)
ax.contour(X1, X2, Z, levels=[0], zdir='z', offset=0, colors='red', linestyles='dashed')
ax.contour(X1, X2, Z, levels=[0], zdir='z', offset=0, colors='blue', linestyles='dashed')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x)$')
plt.title('3D Plot of Nonlinear Programming Problem')
plt.savefig("./Images/optim.pdf")
plt.show()
```
