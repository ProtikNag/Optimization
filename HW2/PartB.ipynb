# Imports

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

# S1 = {x âˆˆ R^3 |x1 = x2, x2 = x3}

```python
t = np.linspace(-10, 10, 100)
S1 = np.column_stack((t, t, t))

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(S1[:, 0], S1[:, 1], S1[:, 2], label='S1')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Sketch of Set S1')
plt.savefig("./Images/s1.pdf")
plt.show()
```

# S2 = {x âˆˆ R^3|x1 + x2 + x3 = 1}

```python
x = np.linspace(0, 1, 1000)
y = np.linspace(0, 1, 1000)
x, y = np.meshgrid(x, y)
z = 1 - x - y

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.8, rstride=100, cstride=100, color='green', label='S2')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Sketch of Set S2')
plt.savefig('./Images/s2.pdf')
plt.show()
```

# S3 = S1 âˆ© S2

```python
t = np.linspace(0, 1, 100)
S3 = np.column_stack((t, t, t))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
x, y = np.meshgrid(x, y)
z = 1 - x - y

ax.plot_surface(x, y, z, alpha=0.8, rstride=100, cstride=100, color='green', label='S2')
ax.plot(S3[:, 0], S3[:, 1], S3[:, 2], color='r', label='S3 (Intersection)')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Sketch of Set S3 (Intersection of S1 and S2)')
plt.savefig("./Images/s3.pdf")
plt.show()
```

# S4 = S1 âŠ• S2

```python
t = np.linspace(-10, 10, 100)
S1 = np.column_stack((t, t, t))

x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
x, y = np.meshgrid(x, y)

# Minkowski XOR
# s3 = s1 xor s2 = (s1 U s2) \ (s1 intersect s2)

S2 = np.column_stack((x.flatten(), y.flatten(), 1 - x.flatten() - y.flatten()))
S4 = np.vstack({tuple(row) for row in np.vstack((S1, S2))})

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(S1[:, 0], S1[:, 1], S1[:, 2], color='b', label='S1')
ax.plot_surface(x, y, 1 - x - y, alpha=0.5, rstride=100, cstride=100, color='c', label='S2')
ax.scatter(S4[:, 0], S4[:, 1], S4[:, 2], color='r', label='S4 (Symmetric Difference)')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Visualization of Symmetric Difference S4 = S1 XOR S2')
plt.show()
```

# f(x) = (x_1)^2 + (x_2)^2

```python
def f(x):
    return x[0]**2 + x[1]**2

x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f([X1, X2])

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('X1-axis')
ax.set_ylabel('X2-axis')
ax.set_zlabel('f(x)')
plt.title('Visualization of the Function $f(x) = x_1^2 + x_2^2$')
plt.savefig("./Images/fx1.pdf")
plt.show()
```

# f(x) = âˆ’x1 ln(x1) âˆ’ x2 ln(x2)

```python
def f(x):
    return -x[0] * np.log(x[0]) - x[1] * np.log(x[1])

x1 = np.linspace(0.01, 2, 100)
x2 = np.linspace(0.01, 2, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f([X1, X2])

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('X1-axis')
ax.set_ylabel('X2-axis')
ax.set_zlabel('f(x)')
plt.title('Visualization of the Function $f(x) = -x_1 \ln(x_1) - x_2 \ln(x_2)$')
plt.savefig("./Images/fx2.pdf")
plt.show()
```

# f(x) = |x1| + |x2|

```python
def f(x):
    return np.abs(x[0]) + np.abs(x[1])

# Create a grid of points
x1 = np.linspace(-2, 2, 100)
x2 = np.linspace(-2, 2, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f([X1, X2])

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('X1-axis')
ax.set_ylabel('X2-axis')
ax.set_zlabel('f(x)')
plt.title('Visualization of the Function $f(x) = |x_1| + |x_2|$')
plt.savefig("./Images/fx3.pdf")
plt.show()
```
