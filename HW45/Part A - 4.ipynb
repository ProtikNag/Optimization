```python
def plot_vectors(vectors, colors, labels, name):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    for v, c, l in zip(vectors, colors, labels):
        ax.quiver(0, 0, 0, v[0], v[1], v[2], color=c, label=l)
        
    ax.view_init(elev=30, azim=25)
    ax.set_xlim([-1, 3])
    ax.set_ylim([-1, 3])
    ax.set_zlim([-1, 3])
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    plt.legend()
    plt.savefig(f"Images/PartA_4_{name}.pdf")
    plt.show()

original_vectors = [A[:, i] for i in range(A.shape[1])]
orthonormal_vectors = [Q[:, i] for i in range(Q.shape[1])]

plot_vectors(original_vectors, ['r', 'g', 'b', 'y'], ['A1', 'A2', 'A3', 'A4'], 'Original Vectors')
plot_vectors(orthonormal_vectors, ['r', 'g'], ['Q1', 'Q2'], 'Orthonormal Vectors')

```
