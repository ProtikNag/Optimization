```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 3 * x**4 - 4 * x**3 + 1

# Define the derivative of the function
def f_prime(x):
    return 12 * x**3 - 12 * x**2

# Define the second derivative of the function
def f_double_prime(x):
    return 36 * x**2 - 24 * x


x = np.linspace(-2, 2, 400)

plt.figure(figsize=(10, 10))
plt.plot(x, f(x), label='f(x) = 3x^4 - 4x^3 + 1', color='blue')

# Mark the horizontal point of inflection where f''(x) = 0
inflection_points = np.roots([12, -12, 0])
plt.scatter(inflection_points, f(inflection_points), color='red', label='Potential Inflection Point')
plt.scatter(0, f(0), color='black', label="Actual Inflection Point")

# Set labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph of f(x) with Point of Inflection')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlim(-4,4)
plt.ylim(-2,5)
plt.legend()
plt.savefig("../Figures/Point of Inflection.pdf")
plt.show()
```

Actual point of inflection x = 0 because f(x) > 1 for x < 0 and f(x) < 1 for x > 0.
