```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog
from scipy.optimize import minimize

import warnings
warnings.filterwarnings("ignore")
```

# Example 3

```python
# Objective function coefficients
c = [-2, -1]

# Inequality constraints matrix
A = [
    [1, 8/3],
    [1, 1],
    [2, 0]
]

# Inequality constraints vector
b = [4, 2, 3]

# Bounds for variables
x0_bounds = (0, None)
x1_bounds = (0, None)

# Solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds], method='highs')

# Extract the optimal solution
x_optimal = result.x

# Plotting the lines
x1_values = np.linspace(0, 2, 500)
x2_values_1 = (4 - x1_values) / (8/3)
x2_values_2 = 2 - x1_values
x2_values_3 = np.ones_like(x1_values) * 3/2

plt.figure(figsize=(12, 8))
plt.plot(x1_values, x2_values_1, label=r'$x_1 + \frac{8}{3}x_2 = 4$')
plt.plot(x1_values, x2_values_2, label=r'$x_1 + x_2 = 2$')
plt.plot(np.ones_like(x1_values) * 3/2, x1_values, label=r'$2x_1 = 3$')

# Plotting the feasible region
x1_values = np.linspace(0, 1.5, 500)
x2_values_1 = (4 - x1_values) / (8/3)
x2_values_2 = 2 - x1_values
x2_values_3 = np.ones_like(x1_values) * 3/2

plt.fill_between(x1_values, 0, np.minimum(np.minimum(x2_values_1, x2_values_2), x2_values_3), color='green', alpha=0.5, label='Feasible Region')

# Mark the optimal solution
plt.plot(x_optimal[0], x_optimal[1], 'ro', label='Optimal Solution')

plt.xlabel('$x_1$', fontsize=12)
plt.ylabel('$x_2$', fontsize=12)
plt.title('Feasible Region and Optimal Solution')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlim(0, 2)
plt.ylim(0, 2)
plt.legend()
plt.savefig("../Figures/Example3.pdf")
plt.show()
```

# Example 7

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (x - 14)**2 + (y - 14)**2

def circle_eq(x, y):
    return (x - 8)**2 + (y - 9)**2 - 49

def line_eq_1(x):
    return 24 - x

# Generate x, y values
x = np.linspace(0, 50, 100)
y = np.linspace(0, 50, 100)
X, Y = np.meshgrid(x, y)

# Plotting the functions
fig, ax = plt.subplots(figsize=(10, 10))

# Contour plot for the first equation
contour = ax.contour(X, Y, f(X, Y), levels=[20, 40, 60, 80], colors='blue', label='f(x, y)')

# Circle plot for the second equation
circle = plt.Circle((8, 9), 7, color='red', fill=False, label='(x - 8)^2 + (y - 9)^2 = 49')
ax.add_patch(circle)

# Line plot for the third equation
line = ax.plot(x, line_eq(x), linestyle='--', color='green', label='x + y = 24')
ax.axvline(x=2, color='purple', linestyle='solid', label='x = 2')
ax.axvline(x=13, color='purple', linestyle='solid', label='x = 13')

# Marking the solution point
ax.scatter(12, 12, color='black', label='Solution Point (11, 10)')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Equations Plot')
ax.set_xlim(0, 30)
ax.set_ylim(0, 30)
ax.legend()

plt.grid(True)
plt.savefig("../Figures/Solution On Boundary.pdf")
plt.show()
```
